```python
import os
import pytz
import pymortar
import datetime
import pandas as pd
import xbos_services_getter
```

```python
meter_data_historical_stub = xbos_services_getter.get_meter_data_historical_stub()
meter_data_historical_stub
```

```python
data_folder = 'data/'

bldg = 'ciee'
point_type='Building_Electric_Meter'
aggregate='RAW'
window= '15m'

tz = pytz.timezone('UTC')
start = tz.localize(datetime.datetime(2018, 2, 8, 0, 0, 0))
end = tz.localize(datetime.datetime(2018, 2, 9, 0, 0, 0))
time_delta = datetime.timedelta(days=1)
```

```python
df = pd.DataFrame()

while start < end:
    
    print('Time: ', start)
    
    temp_end = start + time_delta
    if temp_end > end:
        temp_end = end

    try:
        meter_data = xbos_services_getter.get_meter_data_historical(meter_data_historical_stub, 
                                                                bldg=bldg, 
                                                                start=start, 
                                                                end=temp_end, 
                                                                point_type=point_type, 
                                                                aggregate=aggregate, 
                                                                window=window)

        df = pd.concat([df, meter_data], axis=0)
        print('df.shape: ', df.shape)
        print(df.head())
    
    except Exception as e:
        pass
    
    start = start + time_delta
```
